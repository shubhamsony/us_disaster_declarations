```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df1 = pd.read_csv("us_disaster_declarations.csv")
df2 = pd .read_csv("us_disasters_m5.csv")
  
df1.head()
```

```python
def unique(list1):
 
    unique_list = []
    for x in list1:
        if x not in unique_list:
            unique_list.append(x)
    for x in unique_list:
        print(x)
unique(df['place_code'])
```

```python
df["place_code"].fillna(float(0), inplace = True)
unique(df['place_code'])
```
